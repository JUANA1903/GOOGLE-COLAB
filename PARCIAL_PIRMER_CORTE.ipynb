```python
import pandas as pd # Import the pandas library and assign it to the alias 'pd'

df = pd.read_excel('datos_veterinaria.xlsx') # Now 'pd' is recognized and can be used to call functions like read_excel
```

```python
df.info()
df.describe()
df.head(100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Edad'], bins=20, kde=True)
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig("histograma_edad.png")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df['Especie'], order=df['Especie'].value_counts().index)
plt.title('Distribución de Especie por cliente')
plt.xlabel('Cantidad de especie')
plt.ylabel('Especie')
plt.savefig("barras_Especie.png")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Especie'], y=df['Edad'])
plt.title('Relación entre Especie y Edad')
plt.xlabel('Especie ($)')
plt.ylabel('Puntaje de Edad')
plt.savefig("dispersion_Especie_puntaje.png")
plt.show()
```
